Titanic Passenger Profiles

```python
import pandas as pd

df = pd.read_csv('titanic_passengers.csv')

df.head(25)
```

Sex vs Survival

```python
df.groupby(['Sex', 'Survived']).size().unstack().plot(kind='bar')
```

Passenger Class vs Survival

```python
df.groupby(['Pclass', 'Survived']).size().unstack().plot(kind='bar')
```

Age Distribution by Survival

```python
df[df['Survived']==1]['Age'].plot(kind='hist', alpha=0.5)
df[df['Survived']==0]['Age'].plot(kind='hist', alpha=0.5)
```

Fare Distribution by Survival

```python
df[df['Survived']==1]['Fare'].plot(kind='hist', alpha=0.5)
df[df['Survived']==0]['Fare'].plot(kind='hist', alpha=0.5)
```

Embarked vs Survival

```python
df.groupby(['Embarked', 'Survived']).size().unstack().plot(kind='bar')
```

Family Size vs Survival

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df.groupby(['FamilySize', 'Survived']).size().unstack().plot(kind='bar')
```
